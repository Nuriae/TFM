```python
from qiskit import QuantumCircuit, Aer, assemble
from qiskit.visualization import plot_bloch_multivector

# Crear un circuito cuntico con dos qubits
qc = QuantumCircuit(2)

# Aplicar una puerta CNOT con el qubit 0 como control y el qubit 1 como target
qc.cx(0, 1)

# Dibujar el estado del qubit 0 despus de aplicar la puerta CNOT
backend = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
result = backend.run(qobj).result()
final_state = result.get_statevector()

plot_bloch_multivector(final_state)
```

```python
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
from qiskit import QuantumCircuit

# Vector de estado inicial (0, 1)
initial_state = [0, 1]

# Extender el vector de estado a dos qubits
extended_state = [0, 0, 0,1]

# Convertir el vector de estado en un objeto Statevector
statevector = Statevector(extended_state)

# Crear un circuito cuntico con dos qubits
qc = QuantumCircuit(2)

# Inicializar el estado del circuito con el estado inicial
qc.initialize(statevector, [0, 1])

# Aplicar una puerta CNOT con el qubit 0 como control y el qubit 1 como target
qc.cx(0, 1)

# Obtener el estado final del circuito despus de aplicar la puerta CNOT
final_state = Statevector.from_instruction(qc)

# Dibujar el estado final del vector en la esfera de Bloch
plot_bloch_multivector(final_state)
```

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector

# Crear un circuito cuntico con un qubit
qc = QuantumCircuit(1)

# Aplicar la puerta Pauli-Y al qubit
qc.y(0)

# Obtener el estado final del circuito despus de aplicar la puerta Pauli-Y
final_state = Statevector.from_instruction(qc)

# Dibujar el estado final del vector en la esfera de Bloch
plot_bloch_multivector(final_state)

```
